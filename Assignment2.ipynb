    question :
    building a Fully Connected Feedforward Neural Network (FCFNN) for solving the equation f(x) = 5x^2 +10x -2
    training the FCFNN
    plotting original f(x) and predicted f(x)

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
```

```python
def my_function(x):
    return 5 * (x**2) + 10 * x - 2
```

```python
X = np.linspace(-10, 10, 1000).reshape(-1, 1)
y = my_function(X)
```

```python
def build_model():
    inputs = Input(shape=(1,))
    h1 = Dense(64, activation='relu')(inputs)
    h2 = Dense(128, activation='relu')(h1)
    h3 = Dense(128, activation='relu')(h2)
    h4 = Dense(64, activation='relu')(h3)
    outputs = Dense(1)(h4)
    model = Model(inputs=inputs, outputs=outputs)
    return model
```

```python
model = build_model()
```

```python
model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')
```

```python
history = model.fit(X, y, epochs=500, batch_size=32, verbose=1)
```

```python
y_pred = model.predict(X)
```

```python
plt.figure(figsize=(10,6))
plt.plot(X, y, label="Original f(x)", color="blue")
plt.plot(X, y_pred, label="Predicted f(x)", color="red", linestyle="dashed")
plt.title("Original Function vs Neural Network Prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
